```python
class SistemPakar:
    def __init__(self):
        self.ciri_penyakit = {
            1: [20, 28],  # Mencret
            2: [20, 28],  # Mencret
            3: [28],  # Mencret berdarah
            4: [20, 21, 22, 23, 28],  # Mencret, Muntah, Sakit perut, Darah rendah
            5: [20, 21, 22, 23, 28],  # Mencret, Muntah, Sakit perut, Darah rendah
            6: [21, 28],  # Muntah
            7: [22, 28],  # Sakit perut
            8: [23, 28],  # Darah rendah
            9: [23, 28],  # Darah rendah
            10: [28],  # Lumpuh
            11: [28],  # Demam
            12: [28],  # Demam
            13: [28],  # Demam
            14: [29, 30],  # Makan daging, Makan jamur
            15: [29, 30],  # Makan daging, Makan jamur
            16: [30],  # Makan jamur
            17: [31],  # Makan makanan kaleng
            18: [32],  # Minum susu
            19: [33],  # Minum susu
            20: [20],  # Mencret
            21: [21],  # Muntah
            22: [22],  # Sakit perut
            23: [23],  # Darah rendah
            24: [34],  # Koma
            25: [35, 36],  # Demam, Septicaemia
            26: [35],  # Demam
            27: [36],  # Septicaemia
            28: [37],  # Lumpuh
            29: [14, 15],  # Makan daging, Makan jamur
            30: [16],  # Makan jamur
            31: [17],  # Makan makanan kaleng
            32: [18, 19],  # Minum susu
            33: [33],  # Keracunan Staphylococcus aureus
            34: [34],  # Keracunan jamur beracun
            35: [35],  # Keracunan Salmonellae
            36: [36],  # Keracunan Clostridium botulinum
            37: [37],  # Keracunan Campylobacter
        }

        self.penyakit_ciri = {
            20: [1, 2, 4, 5],  # Mencret
            21: [4, 5, 6],  # Muntah
            22: [4, 7],  # Sakit perut
            23: [4, 8, 9],  # Darah rendah
            24: [8, 10],  # Koma
            25: [4, 5, 9, 11],  # Demam
            26: [4, 8, 11, 12],  # Septicaemia
            27: [4, 13],  # Lumpuh
            28: [1, 2, 3, 4, 5],  # Mencret berdarah
            29: [14, 15],  # Makan daging
            30: [14, 16],  # Makan jamur
            31: [14, 17],  # Makan makanan kaleng
            32: [18, 19],  # Minum susu
            33: [20, 21, 22, 23, 29],  # Keracunan Staphylococcus aureus
            34: [20, 21, 22, 24, 30],  # Keracunan jamur beracun
            35: [20, 21, 22, 25, 26, 29],  # Keracunan Salmonellae
            36: [21, 27, 31],  # Keracunan Clostridium botulinum
            37: [28, 22, 25, 32],  # Keracunan Campylobacter
        }

    def cari_penyakit(self, gejala):
        penyakit_terkait = set()

        # Mencari penyakit yang terkait dengan setiap gejala
        for ciri in gejala:
            penyakit_terkait.update(self.ciri_penyakit.get(ciri, []))

        # Mencari penyakit yang memiliki semua gejala yang terkait
        penyakit_ditemukan = []
        for penyakit in penyakit_terkait:
            if all(ciri in gejala for ciri in self.penyakit_ciri[penyakit]):
                penyakit_ditemukan.append(penyakit)

        return penyakit_ditemukan


# Contoh penggunaan
sistem_pakar = SistemPakar()
gejala_pasien = [1, 2, 4, 5]

hasil_diagnosis = sistem_pakar.cari_penyakit(gejala_pasien)
if hasil_diagnosis:
    print("Penyakit yang mungkin:", hasil_diagnosis)
else:
    print("Tidak ditemukan penyakit yang sesuai dengan gejala yang diberikan.")
```

```python
def diagnosa_gejala(gejala, diagram):
    hasil_diagnosa = set()

    for penyakit, ciri_gejala in diagram.items():
        if all(gejala[gejala_idx] == 1 for gejala_idx in ciri_gejala):
            hasil_diagnosa.add(penyakit)

    return hasil_diagnosa


def main():
    gejala = [
        0
    ] * 37  # Menginisialisasi gejala dengan nilai default (0) untuk semua gejala

    # Set gejala yang sesuai dengan penyakit "Keracunan Staphylococcus aureus"
    gejala[20] = 1  # Mencret
    gejala[21] = 1  # Muntah
    gejala[22] = 1  # Sakit perut
    gejala[23] = 1  # Darah rendah
    gejala[29] = 1  # Makan daging
    # Lanjutkan mengubah nilai gejala sesuai dengan data pasien

    # Diagram infeksi sistem gastro usus
    diagram = {
        33: [20, 21, 22, 23, 29],
        34: [20, 21, 22, 24, 30],
        35: [20, 21, 22, 25, 26, 29],
        36: [21, 27, 31],
        37: [28, 22, 25, 32],
    }

    hasil_diagnosa = diagnosa_gejala(gejala, diagram)

    print("Hasil diagnosa:")
    for penyakit in hasil_diagnosa:
        print(f"- Penyakit {penyakit}: {diagram[penyakit]}")


if __name__ == "__main__":
    main()
```

```python
gejala = {
    "Mencret": [1, 2, 4, 5],
    "Muntah": [4, 5, 6],
    "Sakit perut": [4, 7],
    "Koma": [8, 10],
    "Demam": [4, 5, 9, 11],
    "Sepsis": [4, 8, 11, 12],
    "Lumpuh": [4, 13],
    "Darah rendah": [4, 8, 9],
    "Mencret berdarah": [1, 2, 3, 4, 5],
}

penyebab = {
    "Keracunan Staphylococcus aureus": [14, 15],
    "Keracunan jamur beracun": [14, 16],
    "Keracunan Salmonellae": [14, 15],
    "Keracunan Clostridium botulinum": [14, 17],
    "Keracunan Campylobacter": [18, 19],
}


def diagnosis(gejala_pasien):
    diagnosis = {}
    for jenis_keracunan, daftar_gejala in penyebab.items():
        skor = 0
        for gejala in gejala_pasien:
            if gejala in daftar_gejala:
                skor += 1
        diagnosis[jenis_keracunan] = skor

    diagnosis_tertinggi = max(diagnosis, key=diagnosis.get)
    return diagnosis_tertinggi


while True:
    gejala_pasien = []
    while True:
        input_gejala = input("Masukkan kode gejala (pisahkan dengan koma): ")
        gejala_list = input_gejala.split(",")

        for item in gejala_list:
            # Check if the item is a digit (number)
            if item.isdigit():
                # Convert the digit to an integer and check if it's in the gejala dictionary
                if int(item) in sum(gejala.values(), []) + sum(penyebab.values(), []):
                    gejala_pasien.append(int(item))
                else:
                    print(f"Kode gejala {item} tidak valid. Silakan coba lagi.")
            else:
                print(
                    f"Input '{item}' bukan angka. Masukkan kode gejala yang valid (pisahkan dengan koma)."
                )

        if len(gejala_pasien) > 0:
            break

    diagnosis_tertinggi = diagnosis(gejala_pasien)
    print(f"Diagnosis: {diagnosis_tertinggi}")

    # Ask the user if they want to continue
    lanjut = input("Apakah Anda ingin mencoba lagi? (y/n): ")
    if lanjut.lower() != "y":
        break
```

```python
# Contoh 1: Keracunan Staphylococcus aureus
gejala_pasien = [1,2,4,6,14,15]  # Gejala: Mencret, Muntah, Sakit perut, Darah rendah, Makan daging
print(diagnosis(gejala_pasien))  # Output: Keracunan Staphylococcus aureus

# Contoh 2: Keracunan jamur beracun
gejala_pasien = [1,2,4,5,8,10,14,16]  # Gejala: Mencret, Muntah, Sakit perut, Koma, Makan jamur
print(diagnosis(gejala_pasien))  # Output: Keracunan jamur beracun

# Contoh 3: Keracunan Salmonellae
gejala_pasien = [1,2,4,5,9,11,14,15]  # Gejala: Mencret, Muntah, Sakit perut, Demam, Septicaemia, Makan daging
print(diagnosis(gejala_pasien))  # Output: Keracunan Salmonellae

# Contoh 4: Keracunan Clostridium botulinum
gejala_pasien = [4,5,6,13,14,17]  # Gejala: Muntah, Lumpuh, Makan makanan kaleng
print(diagnosis(gejala_pasien))  # Output: Keracunan Clostridium botulinum

# Contoh 5: Keracunan Campylobacter
gejala_pasien = [1,2,3,4,5,7,9,11,18,19]  # Gejala: Mencret berdarah, Sakit perut, Demam, Minum susu
print(diagnosis(gejala_pasien))  # Output: Keracunan Campylobacter
```

```python
class Pasien:
    def __init__(self, gejala):
        self.gejala = gejala


def cek_keracunan(pasien):
    if set([1, 2, 4, 6]).issubset(pasien.gejala) and set([14, 15]).issubset(
        pasien.gejala
    ):
        return "Keracunan Staphylococcus aureus"
    elif set([1, 2, 4, 5]).issubset(pasien.gejala) and set([14, 16]).issubset(
        pasien.gejala
    ):
        return "Keracunan jamur beracun"
    elif set([1, 2, 4, 5]).issubset(pasien.gejala) and set([14, 15]).issubset(
        pasien.gejala
    ):
        return "Keracunan Salmonellae"
    elif set([4, 5, 6]).issubset(pasien.gejala) and set([14, 17]).issubset(
        pasien.gejala
    ):
        return "Keracunan Clostridium botulinum"
    elif set([1, 2, 3, 4, 5]).issubset(pasien.gejala) and set([18, 19]).issubset(
        pasien.gejala
    ):
        return "Keracunan Campylobacter"
    else:
        return "Tidak dapat menentukan jenis keracunan"


# Contoh 1: Keracunan Staphylococcus aureus
gejala_pasien = [1,2,4,6,14,15]  # Gejala: Mencret, Muntah, Sakit perut, Darah rendah, Makan daging
pasien = Pasien(gejala_pasien)
print(cek_keracunan(pasien))  # Output: Keracunan Staphylococcus aureus

# Contoh 2: Keracunan jamur beracun
gejala_pasien = [1,2,4,5,8,10,14,16]  # Gejala: Mencret, Muntah, Sakit perut, Koma, Makan jamur
pasien = Pasien(gejala_pasien)
print(cek_keracunan(pasien))  # Output: Keracunan jamur beracun

# Contoh 3: Keracunan Salmonellae
gejala_pasien = [1,2,4,5,9,11,14,15]  # Gejala: Mencret, Muntah, Sakit perut, Demam, Septicaemia, Makan daging
pasien = Pasien(gejala_pasien)
print(cek_keracunan(pasien))  # Output: Keracunan Salmonellae

# Contoh 4: Keracunan Clostridium botulinum
gejala_pasien = [4,5,6,13,14,17]  # Gejala: Muntah, Lumpuh, Makan makanan kaleng
pasien = Pasien(gejala_pasien)
print(cek_keracunan(pasien))  # Output: Keracunan Clostridium botulinum

# Contoh 5: Keracunan Campylobacter
gejala_pasien = [1,2,3,4,5,7,9,11,18,19]  # Gejala: Mencret berdarah, Sakit perut, Demam, Minum susu
pasien = Pasien(gejala_pasien)
print(cek_keracunan(pasien))  # Output: Keracunan Campylobacter
```
